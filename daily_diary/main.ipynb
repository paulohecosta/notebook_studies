## Diario de Bordo

Diario de bordo para acompanhamento de pessoas e projetos

```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diary = pd.read_csv('arqs.csv')
diary
```

```python
cross_tab = pd.crosstab(diary['arq'], diary['sigla'])

plt.figure(figsize=(5, 3))
sns.heatmap(cross_tab, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap entre Arq e Sigla')
plt.show()
```

```python
cross_tab = pd.crosstab(diary['arq'], diary['priority'])

priority_order = ['A', 'M', 'B']
cross_tab = cross_tab[priority_order]

plt.figure(figsize=(5, 3))
sns.heatmap(cross_tab, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap entre Arq e Priority')
plt.show()
```

```python
cross_tab = pd.crosstab(diary['arq'], diary['chap'])

plt.figure(figsize=(5, 3))
sns.heatmap(cross_tab, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap entre Arq e Chapters')
plt.show()
```

```python
# Criando uma pivot table para usar no heatmap
pivot_table = diary.pivot_table(index='arq', columns='sigla', values='dif_arq', aggfunc='first')

# Criando o heatmap usando a biblioteca seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='.1f')  # Usando formato de ponto flutuante com uma casa decimal
plt.title('Heatmap entre Arq e Diferena de Arq')
plt.show()
```
